```python
import requests
```

```python
url = "https://en.wikipedia.org/wiki/Disneyland"
```

```python
r = requests.get(url)
```

```python
print(r.text)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(r.text , 'html.parser')
```

```python
print(soup)
```

```python
heading = soup.find('h1').text
```

```python
heading
```

```python
print(soup.text.replace('\n\n', ''))
```

```python
for p in soup.find_all('p'):
    print(p.text)
    print('--'*10)
```

```python
corpus = ''

for p in soup.find_all('p'):
    corpus += p.text
    corpus += '\n'
    
corpus = corpus.strip()    
```

```python
print(corpus)
```

```python
for i in range(2,137):
    corpus= corpus.replace('['+ str(i)+']','')
```

```python
print(corpus)
```

# Creating a txt file 

```python
fd = open(heading + '.txt', 'w',encoding='utf-8')
fd.write(corpus)
fd.close()   
```
